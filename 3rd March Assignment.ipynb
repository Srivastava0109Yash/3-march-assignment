### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a popular Python library used for data visualization. Here are five types of plots that can be created using Seaborn along with their uses:

#### scatterplot
#### histplot
#### relpolt
#### pairplot
#### catplot

### Use:

__1.  scatterplot__: This plot is used to show the relationship between two continuous variables. It can help in identifying any patterns or trends in the data

__2.  histplot__:  This plot is used to create a histogram of a variable or a dataset. A histogram is a type of plot that displays the distribution of continuous numerical data.

__3.  relpolt__: This plot is used to create a relational plot, it is a flexible function that can craete different types of plots depending on the input data and type of plot to be created.

__4.  pairplot__: This plot is used to create a pairwise plot for a dataset. It creates scatter plots between all pairs of variables in the dataset and histograms along the diagonal to show the distribution of each variable.

__5.  catplot__: This is used to create several types of categorical plots such as bar plots, box plots, strip plots, violin plots, etc. It allows the user to specify the type of plot, the variables to be plotted, and the data to be used. can also be used to group data by one or more variables and to apply statistical functions to the groups

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns

fmri=sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x=fmri['timepoint'],y=fmri['signal'])
```

```python
sns.lineplot(x=fmri['timepoint'],y=fmri['signal'],data=fmri,hue='event')
```

```python
sns.lineplot(x=fmri['timepoint'],y=fmri['signal'],data=fmri,style='region')
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights=sns.load_dataset('flights')
```

```python
flights
```

```python
flights = flights.pivot("year","month", "passengers")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
sns.heatmap(data=flights,annot=True,robust=True)
```
